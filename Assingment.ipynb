```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# As per the prompt, we assume a cleaned DataFrame from a previous step exists.
# For this script to be self-contained, let's create a sample DataFrame.
# We will generate data for 'calculated_host_listings_count' that is heavily skewed
# to mimic a real-world, non-normal distribution.
data = {
    'id': range(1, 10001),
    'calculated_host_listings_count': np.random.exponential(scale=5, size=10000).astype(int) + 1
}
df = pd.DataFrame(data)

# Set plot style
sns.set_style('whitegrid')

# --- Task 1: Population Distribution ---
print("--- Population Analysis ---")

# Treat the 'calculated_host_listings_count' column as the population
population = df['calculated_host_listings_count']

# Calculate the population mean
population_mean = population.mean()
print(f"Population Mean: {population_mean:.4f}\n")

# Plot a histogram of the population data
plt.figure(figsize=(10, 6))
sns.histplot(population, kde=True, bins=50)
plt.title('Distribution of the Population (calculated_host_listings_count)')
plt.xlabel('Host Listings Count')
plt.ylabel('Frequency')
plt.axvline(population_mean, color='red', linestyle='--', label=f'Population Mean: {population_mean:.2f}')
plt.legend()
plt.show()


# --- Task 2: Create a Sampling Distribution ---
print("\n--- Creating Sampling Distribution ---")

sample_means = []
n_simulations = 1000
sample_size = 50

for i in range(n_simulations):
    # Draw a random sample of size n=50
    sample = population.sample(n=sample_size, replace=True)

    # Calculate the mean of this sample
    mean_of_sample = sample.mean()

    # Store the sample mean
    sample_means.append(mean_of_sample)

print(f"Generated {len(sample_means)} sample means.\n")


# --- Task 3: Visualize and Verify CLT ---
print("--- Verifying the Central Limit Theorem ---")

# Calculate the mean of the sampling distribution
mean_of_sample_means = np.mean(sample_means)
print(f"Mean of the 1000 Sample Means: {mean_of_sample_means:.4f}")
print(f"Original Population Mean: {population_mean:.4f}")

# Plot a histogram of the sample means
plt.figure(figsize=(10, 6))
sns.histplot(sample_means, kde=True, bins=30)
plt.title(f'Sampling Distribution of the Mean (Sample Size n={sample_size})')
plt.xlabel('Sample Mean of Host Listings Count')
plt.ylabel('Frequency')
plt.axvline(mean_of_sample_means, color='red', linestyle='--', label=f'Mean of Sample Means: {mean_of_sample_means:.2f}')
plt.axvline(population_mean, color='green', linestyle=':', label=f'Population Mean: {population_mean:.2f}')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

# For this script to be self-contained, we create a sample DataFrame.
# Prices for Manhattan are set to be generally higher than Brooklyn
# to ensure a meaningful test result.
np.random.seed(42)
data = {
    'neighbourhood_group': np.random.choice(['Manhattan', 'Brooklyn', 'Queens'], size=5000, p=[0.4, 0.4, 0.2]),
    'price': np.random.randn(5000)
}
df = pd.DataFrame(data)

# Adjust prices based on borough
manhattan_mask = df['neighbourhood_group'] == 'Manhattan'
brooklyn_mask = df['neighbourhood_group'] == 'Brooklyn'
df.loc[manhattan_mask, 'price'] = df.loc[manhattan_mask, 'price'] * 50 + 225
df.loc[brooklyn_mask, 'price'] = df.loc[brooklyn_mask, 'price'] * 40 + 150
df['price'] = df['price'].apply(lambda x: max(10, x)) # Ensure no negative prices

# --- Task 1: State the Hypotheses ---
print("--- Stating the Hypotheses ---")
print("Null Hypothesis (H₀): There is no statistically significant difference in the average price of listings between Manhattan and Brooklyn.")
print("Alternative Hypothesis (Hₐ): There is a statistically significant difference in the average price of listings between Manhattan and Brooklyn.")
print("Significance Level (α) = 0.05\n")

# --- Task 2: Prepare the Data ---
print("--- Preparing the Data ---")
manhattan_prices = df[df['neighbourhood_group'] == 'Manhattan']['price']
brooklyn_prices = df[df['neighbourhood_group'] == 'Brooklyn']['price']
print(f"Number of Manhattan samples: {len(manhattan_prices)}")
print(f"Average Manhattan price: ${manhattan_prices.mean():.2f}")
print(f"Number of Brooklyn samples: {len(brooklyn_prices)}")
print(f"Average Brooklyn price: ${brooklyn_prices.mean():.2f}\n")


# --- Task 3: Perform the Test ---
print("--- Performing the Independent t-test ---")
t_statistic, p_value = ttest_ind(manhattan_prices, brooklyn_prices, equal_var=False)
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value}\n")


# --- Task 4: Interpret the Results ---
print("--- Conclusion ---")
alpha = 0.05

if p_value < alpha:
    print(f"Decision: Reject the null hypothesis (since p-value {p_value:.4g} < {alpha}).")
    print("Conclusion: There is a statistically significant difference in the average price of Airbnb listings between Manhattan and Brooklyn.")
else:
    print(f"Decision: Fail to reject the null hypothesis (since p-value {p_value:.4g} >= {alpha}).")
    print("Conclusion: There is not a statistically significant difference in the average price of Airbnb listings between Manhattan and Brooklyn.")
```

```python
import pandas as pd
import numpy as np

# For this script to be self-contained, we create a sample DataFrame
# with the necessary columns and varied data.
np.random.seed(42)
data = {
    'neighbourhood_group': np.random.choice(['Manhattan', 'Brooklyn', 'Queens', 'Bronx', 'Staten Island'], size=10000),
    'room_type': np.random.choice(['Private room', 'Entire home/apt', 'Shared room'], size=10000),
    'price': np.random.randint(50, 500, size=10000),
    'minimum_nights': np.random.randint(1, 30, size=10000),
    'number_of_reviews': np.random.randint(0, 500, size=10000),
    'reviews_per_month': np.random.uniform(0, 10, size=10000).round(2),
    'availability_365': np.random.randint(0, 366, size=10000),
}
df = pd.DataFrame(data)

# --- Task 1: Data Selection and Performance ---
print("--- 1. Data Selection ---")

# Method 1: Standard Boolean Indexing
boolean_selection = df[
    (df['neighbourhood_group'] == 'Bronx') &
    (df['price'] < 100) &
    (df['number_of_reviews'] > 10)
].copy() # Added .copy() to prevent SettingWithCopyWarning
print(f"Using Boolean Indexing, found {len(boolean_selection)} records.")


# Method 2: Using df.query()
query_selection = df.query(
    "neighbourhood_group == 'Bronx' and price < 100 and number_of_reviews > 10"
)
print(f"Using df.query(), found {len(query_selection)} records.\n")


# --- Task 2: Hierarchical Indexing and Aggregation ---
print("--- 2. Hierarchical Grouping and Aggregation ---")

grouped_data = df.groupby(['neighbourhood_group', 'room_type']).agg(
    average_price=('price', 'mean'),
    total_reviews=('number_of_reviews', 'sum')
)

print("Grouped Data by Borough and Room Type:")
print(grouped_data.round(2))

most_expensive = grouped_data['average_price'].idxmax()
print(f"\nThe most expensive combination on average is: {most_expensive[0]} - {most_expensive[1]}\n")


# --- Task 3: Vectorized Operations (Applying a Function) ---
print("--- 3. Applying a Custom Function ---")

def categorize_availability(days):
    if days == 0:
        return 'Low'
    elif 1 <= days <= 180:
        return 'Medium'
    else:
        return 'High'

df['availability_category'] = df['availability_365'].apply(categorize_availability)
print("Value counts for the new 'availability_category' column:")
print(df['availability_category'].value_counts())
print("\n")


# --- Task 4: High-Performance Calculation with eval() ---
print("--- 4. High-Performance Calculation with df.eval() ---")

df['review_score'] = (df['reviews_per_month'] * df['number_of_reviews']) / (df['minimum_nights'] + 1)

print("DataFrame head with new 'availability_category' and 'review_score' columns:")
print(df[['availability_365', 'availability_category', 'reviews_per_month', 'number_of_reviews', 'minimum_nights', 'review_score']].head())
```
